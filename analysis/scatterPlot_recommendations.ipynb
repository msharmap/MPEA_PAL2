```python
## Misc functions for data editing
# For HEA dataset
import os
import numpy as np
import pandas as pd
import pymatgen as mg

# Plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Tick parameters
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['xtick.major.size'] = 5
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['xtick.minor.size'] = 5
plt.rcParams['xtick.minor.width'] = 1
plt.rcParams['ytick.major.size'] = 5
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['ytick.minor.size'] = 5
plt.rcParams['ytick.minor.width'] = 1

plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 15
plt.rcParams['legend.fontsize'] = 15

datafile_path = '/Users/maitreyeesharma/WORKSPACE/PostDoc/EngChem/Space@Hopkins_HEA/dataset/'
unique_alloys_training2 = pd.read_csv(datafile_path+'second_pass/Recommendations_Oct7_EBG.csv')
unique_alloys_training3 = pd.read_csv(datafile_path+'third_pass/Recommendations_APR24_ebg.csv')

unique_alloys_training1 = pd.DataFrame()
unique_alloys_training1['Hardness_Value_predicted']=[868,627]
unique_alloys_training1['Hardness_Value_measured']=[733,763]
```

```python
## Scatter plot of predicted versus measured HV
plt.plot([300,1100],[300,1100],'-k')
plt.scatter(unique_alloys_training1['Hardness_Value_predicted'],unique_alloys_training1['Hardness_Value_measured'],color='r')
plt.scatter(unique_alloys_training2['Hardness_Value_predicted'],unique_alloys_training2['Hardness_Value_measured'],color='r')
plt.scatter(unique_alloys_training3['Hardness_Value_predicted'],unique_alloys_training3['Hardness_Value_measured'],color='r')

plt.xlim([600,1100])
plt.ylim([600,1100])
```

```python
## Percentage error of predicted HV
error1 = np.mean(np.abs(unique_alloys_training1['Hardness_Value_predicted']
                      -unique_alloys_training1['Hardness_Value_measured'])
                /unique_alloys_training1['Hardness_Value_measured'])
error2 = np.mean(np.abs(unique_alloys_training2['Hardness_Value_predicted']
                      -unique_alloys_training2['Hardness_Value_measured'])
                /unique_alloys_training2['Hardness_Value_measured'])
error3 = np.mean(np.abs(unique_alloys_training3['Hardness_Value_predicted']
                      -unique_alloys_training3['Hardness_Value_measured'])
                /unique_alloys_training3['Hardness_Value_measured'])               
error = 1.0/3.0*(error1+error2+error3)            
               
               
```
